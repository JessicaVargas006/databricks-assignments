```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('data_processing'). getOrCreate()
```

```python
import pyspark.sql.functions as F
from pyspark.sql.types import *
```

```python
schema=StructType().add("user_id","string").add("country","string").add("browser","string").add("OS",'string').add("age", "integer")
```

```python
df=spark.createDataFrame([("A203",'India',"Chrome","WIN",
33),("A201",'China',"Safari","MacOS",35),("A205",'UK',"Mozilla",
"Linux",25)],schema=schema)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
df_na=spark.createDataFrame([("A203",None,"Chrome","WIN",
33),("A201",'China',None,"MacOS",35),("A205",'UK',"Mozilla",
"Linux",25)],schema=schema)
```

```python
df_na.show()
```

```python
df_na.fillna('0').show()
```

```python
df_na.fillna( { 'country':'USA', 'browser':'Safari' } ).show()
```

```python
df_na.na.drop().show()
```

```python
df_na.na.drop(subset='country').show()
```

```python
df_na.replace("Chrome","Google Chrome").show()
```

```python
df_na.drop('user_id').show()
```

## What We Learned 

- Start Spark with **SparkSession**; it’s the main handle (`spark`).
- Define an explicit **schema** so column types are correct.
- Create a **DataFrame** from rows using that schema.
- Use **`printSchema()`** to check types and **`show()`** to preview data.
- Handle missing data:
  - **`fillna()`** to fill nulls (global or per column).
  - **`na.drop()`** to drop rows with nulls (all or by subset).
- Change values with **`replace(old, new)`**.
- Remove columns with **`drop('col')`**.
- Spark transforms return **new DataFrames** (original isn’t changed).
