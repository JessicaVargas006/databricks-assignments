```python
# In-class activity 8 starts here

# Create a dataframe in a different notebook named activity_8
# Attach a new cluster to the notebook
# Create the fields below to populate the data frame
# 1001 basketball court Kendall 2500 inspection passed
# 1002 baseball field Kendall 600 fence to be inspected
# 1003 library Hialeah 240
# 1004 Gallery North 120 Open to the public
# 1005 Idea Center 200 Events scheduled

from pyspark.sql.types import StructType, StructField, IntegerType, StringType
from pyspark.sql import functions as F

schema = StructType([
    StructField("id", IntegerType(), False),
    StructField("facility", StringType(), True),
    StructField("campus", StringType(), True),
    StructField("capacity", IntegerType(), True),
    StructField("description", StringType(), True)
])

data = [
    (1001, "basketball court", "Kendall", 2500, "inspection passed"),
    (1002, "baseball field",   "Kendall",  600, "fence to be inspected"),
    (1003, "library",          "Hialeah",  240, None),                # NULL description
    (1004, "Gallery",          "North",    120, "Open to the public"),
    (1005, "Idea Center",      None,       200, "Events scheduled")   # NULL campus
]

df = spark.createDataFrame(data, schema)
display(df)  # or df.show(truncate=False)
```

```python
# Show the dataframe if we replaced Null values with 0. Would that make sense? Explain

df_zero = df.na.fill(0)   # fills numeric nulls (e.g., capacity) with 0; text nulls remain
display(df_zero)
```

Explanation:
Replacing NULL with 0 only makes sense for numeric columns when 0 is a real value.
It is not appropriate for unknown text values (e.g., campus/description).

```python
# Show the dataframe if we filled the null values for the fields campus with "District" and description with "Not Available".

df_fill = df.na.fill({"campus": "District", "description": "Not Available"})
display(df_fill)
```

```python
# Show how the existing dataframe would look like with rows including null values being dropped

df_drop_any = df.na.drop("any") 
display(df_drop_any)
```

```python
# Show how the existing dataframe would look like with rows including null values for 'Description' being dropped

df_drop_desc = df.na.drop(subset=["description"])
display(df_drop_desc)
```

```python
# Show how the existing dataframe would look like with rows including null values for campus being dropped

df_drop_campus = df.na.drop(subset=["campus"])
display(df_drop_campus)
```

```python
#Show how the existing dataframe would look like if we replaced "baseball field" with baseball
df_replace = df.replace("baseball field", "baseball", subset=["facility"])
display(df_replace)
```

```python
#Show how the existing dataframe would look like if we dropped the description column
df_no_description = df.drop("description")
display(df_no_description)
```

**Activity 8 Summary**

- We recreated a DataFrame (id, facility, campus, capacity, description) and purposely included some NULL values (missing campus/description).
- We tried fillna(0) and saw it only makes sense for numeric columns (e.g., capacity). Using 0 for unknown text doesn’t make sense.
- We filled text NULLs with labels using na.fill({"campus":"District","description":"Not Available"}) to keep reports readable without losing meaning.
- We practiced dropping rows with missing data:
- na.drop("any") removed any row that had any NULL.
- na.drop(subset=["description"]) only removed rows missing description.
- na.drop(subset=["campus"]) only removed rows missing campus.
- We replaced a value ("baseball field" → "baseball") using replace / regexp_replace.
- We dropped a column (description) with drop("description").
- Throughout, we used display() to compare results and understand how each step changed the table.
**Key takeaways:**
- Choose fill vs. drop based on the question you’re answering; labels are better than 0 for unknown text.
- Use subset to target specific columns when dropping.
- Small cleanups (replace/drop) make downstream analytics simpler and less error-prone.

**One-liner:** Activity 8 was about practical data cleaning in Spark—knowing when to fill, drop, or replace missing or messy values and seeing how each choice affects the DataFrame.
