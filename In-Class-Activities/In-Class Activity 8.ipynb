```python
from pyspark.sql.functions import udf
```

```python

df.select(['ScopeofWork']).show()
```

```python
def Category(ScopeofWork):
  if (ScopeofWork =="REMODELING/REPAIRS") or (ScopeofWork=="BUILDING ROOFING") or (ScopeofWork=="NEW CONSTRUCTION"):
   return "Construction"
  elif ScopeofWork== "SIGN":
   return "Marketing"
  elif ((ScopeofWork== "PLUMBING") or (ScopeofWork== "ELECTRICAL") or (ScopeofWork=="MECHANICAL") ) :
   return "Maintenance"
  else:
   return "Attention"
```

```python
Category_udf=udf(Category,StringType())
```

```python
df=df.withColumn('Category',Category_udf(df['ScopeofWork']))
```

```python
df.select('ScopeofWork','Category').show()
```

```python
df.groupby("Category").count().show()
```
