```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('data_preprocessing'). getOrCreate()
```

```python
import pyspark.sql.functions as F
```

```python
from pyspark.sql.types import *
```

```python
schema=StructType().add("student_id","string").add("major","string").add("campus", "string").add("cumulative_credits",'integer').add("semester_credits", "integer")
```

```python
df=spark.createDataFrame([("K201","Math","West",64,
13),("K202","Biology","Kendall",31,6),("A203","CSI","Hialeah",
82,15),("K204","Veterinary","Kendall",61,12),("K205","Education","Padron",101,10),("K206","History","North",55,16),("K207","Data Analytics","Kendall",56,8),
("K208","Business","Homestead",59,12),("K209","IT","Wolfson",72,4)],schema=schema)
```

```python
df.printSchema()
```
