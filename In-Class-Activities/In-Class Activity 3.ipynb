```python
# 1. Load the "quiz_2" table as a Spark DataFrame
from pyspark.sql import SparkSession
from pyspark.sql.functions import when

# Create or get existing Spark session
spark = SparkSession.builder.appName("Quiz2").getOrCreate()

# Load the quiz_2 table
# (if it's a saved table inside your Spark catalog)
df = spark.table("quiz_2")

# OR if your data is in a CSV file (uncomment below instead)
# df = spark.read.csv("quiz2.csv", header=True, inferSchema=True)

# 2. Create a new column "high_income" that is 1 if "income" > 50000, otherwise 0
df = df.withColumn("high_income", when(df["income"] > 50000, 1).otherwise(0))

# 3. Display the first 10 rows of the resulting DataFrame
df.show(10)
```
