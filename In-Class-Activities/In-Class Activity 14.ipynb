```python
#The first step is to create the SparkSession object, in order to use Spark.
from pyspark.sql import SparkSession
#Import the pyspark sql functions 
import pyspark.sql.functions as F
#Import all the sql data types and functions from pyspark
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
#Create a SparkSession called 'chi_sq_test'
spark = SparkSession.builder.appName('chi_sq_test').getOrCreate()
```

```python
#Read the csv file into a Spark DataFrame
df1 = spark.read.csv('/Workspace/Users/jessica.vargas006@mymdc.net/chi_sq2.csv',
                     header = 'True',
                     inferSchema = 'True')
#Show the dataframe                     
df1.show()
```

```python
#Count the number of records
df1.count()
```

```python
#Import the StringIndexer from the pyspark libraries
from pyspark.ml.feature import StringIndexer
```

```python
#Create a String Indexer object called marital_indexer
marital_indexer = StringIndexer(inputCol = 'marital',
                                outputCol = 'marital_num').fit(df1)
```

```python
#Apply the marital_indexer on the original df1 to a new dataframe 'df'
df = marital_indexer.transform(df1)
df.show()
```

```python
#Import the OneHotEncoder from the pyspark libraries
from pyspark.ml.feature import OneHotEncoder
```

```python
#Drop the new column if it exists
df = df.drop('marital_vect')
#Create a OneHotEncoder object called 'marital_encoder' that encodes the input col to output
marital_encoder = OneHotEncoder(inputCol = 'marital_num',
                                outputCol = 'marital_vect')
marital_encoder.setDropLast(False)
ohe = marital_encoder.fit(df)                                
df = ohe.transform(df)
#display the transformed vector
display(df)
```

```python
#drop the original columns
df = df.drop('housing_num')
#Create a StringIndexer object called 'housing_indexer'
housing_indexer = StringIndexer(inputCol = 'housing',
                                outputCol = 'housing_num').fit(df)
#Apply the transformation
df = housing_indexer.transform(df)
#Show the dataframe
df.show()
```

```python
#Drop the new columnds if they exist
df = df.drop('housing_vect')
#Create a OneHotEncoder object called 'housing_encoder'
housing_encoder = OneHotEncoder(inputCol = 'housing_num',
                                outputCol = 'housing_vect')
housing_encoder.setDropLast(False)
ohe = housing_encoder.fit(df)
#Apply the transformation
df = ohe.transform(df)
#display the transformed vector
display(df)
```

```python
#Show the dataframe
df.show()
```

```python
#Import the VectorAssembler from the pyspark libraries
from pyspark.ml.feature import VectorAssembler
#Drop the new column if it exists
df = df.drop('features')
#Create a VectorAssembler object called 'vectAssembler'
#This will assemble the columns 'marital_vect' and 'housing_vect' into a single vector column 'features'
vectAssembler = VectorAssembler(inputCols = ['marital_vect', 'housing_vect'],
                               outputCol = 'features')
df = vectAssembler.transform(df)
df.select('marital_vect', 'features', 'housing_vect').show(10)
```

```python
#Show the dataframe
df.show()
```

```python
#Import the ChiSquareTest from the pyspark libraries
from pyspark.ml.stat import ChiSquareTest
```

```python
#Create a ChiSquareTest object called chi_sq
chi_sq = ChiSquareTest.test(df, 'features', 'marital_num').head()
```

```python
print(chi_sq)
```

```python
print('pValues: ' + str(chi_sq.pValues))
```

### Reflection Questions – Chi-square Activity

**1. What Databricks libraries did we use for this problem?**  
We used PySpark libraries provided in Databricks, mainly:
- `pyspark.sql` for working with DataFrames and the `SparkSession`
- `pyspark.sql.functions` (imported as `F`) for column transformations
- `pyspark.ml.feature` for feature engineering (in our solution we used `VectorAssembler`, and conceptually this is also where tools like `StringIndexer`/`OneHotEncoder` live)
- `pyspark.ml.stat` for statistical methods, specifically `ChiSquareTest`

---

**2. When do we use the Chi-square test method?**  
We use the Chi-square test when we want to check **whether two categorical variables are independent**.  
In machine learning, it is often used in classification problems to test if a categorical feature is statistically related to the target label or not.

---

**3. What are the features in this problem?**  
The features are the customer characteristics:
- `marital` (encoded as `marital_num`)
- `housing` (encoded as `housing_num`)

These two variables are used as predictors in the Chi-square test.

---

**4. What were the transformations applied?**  
- Loaded the CSV file into a Spark DataFrame.  
- Converted the categorical variables to numeric codes:
  - `marital` → `marital_num` (e.g., *single* = 0, *married* = 1)  
  - `housing` → `housing_num` (e.g., *no* = 0, *yes* = 1)  
- Used `VectorAssembler` to combine `marital_num` and `housing_num` into a single `features` vector column so they could be passed to `ChiSquareTest`.

---

**5. What is the target variable in this problem?**  
The target variable (label we are trying to explain) is:

- `label`

This column is binary (0/1) and represents the outcome we are testing against the features.

---

**6. What is the result of the Chi-square test? Interpret it.**  
The Chi-square test produced the following results:

- For **`marital_num` vs `label`**:  
  - χ² ≈ **0.055**, p-value ≈ **0.815**  
  - The p-value is much larger than 0.05, so we **fail to reject the null hypothesis** of independence.  
  - Interpretation: there is **no evidence of a significant relationship** between marital status and the label.

- For **`housing_num` vs `label`**:  
  - χ² ≈ **3.47**, p-value ≈ **0.063**  
  - The p-value is slightly above 0.05, so at the 5% significance level we also **fail to reject independence**.  
  - Interpretation: housing shows a **weak or borderline association**, but it is **not statistically significant** at α = 0.05. There is only limited evidence that housing status is related to the label.

Overall, the Chi-square test suggests that neither marital status nor housing status has a strong, statistically significant association with the target variable in this dataset.
