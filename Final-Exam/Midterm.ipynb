```python
import pyspark.sql.functions as F
from pyspark.sql.types import *
```

```python
#Now you will create a dataframe using the following information about the customers of a local bank.
```

```python
#customer_id, cust_fname, cust_lname,cash_flow, account_type
schema = StructType([
    StructField("customer_id", IntegerType(), False), 
    StructField("cust_fname", StringType(), True),
    StructField("cust_lname", StringType(), True),
    StructField("cash_flow", IntegerType(), True),
    StructField("account_type", StringType(), True)
])
```

```python
#Make sure that customer_id does not take null values. 
data = [
    (101, "Jorge", "Perez", 2500, "Personal"),
    (102, None, "Jones", 60000, "Business"),
    (103, "Kelly", "Maine", 9200, "Personal"),
    (104, "Gabriella", "Martin", 109200, "Business"),
    (105, "Michelle", "Abrams", 29456, None)
]
```

```python
from pyspark.sql import Row
#Show the dataframe if we replaced Null values with 0. Would that make any sense? Explain.
# Sample data with nulls
data = [
    Row(cust_fname=None, account_type=None, cash_flow=100),
    Row(cust_fname="Alice", account_type=None, cash_flow=None),
    Row(cust_fname=None, account_type="Savings", cash_flow=200)
]

df = spark.createDataFrame(data)

print("*** Replacing Nulls with 0 ***")
display(
    df.fillna("0")
)

# EXPLANATION: 
# It does NOT make sense to replace nulls with 0 in this context. 
# 1. "0" is not a valid name or account type, so it corrupts the data quality.
# 2. For statistical analysis (like cash_flow), filling with 0 (if that was missing) might skew the average significantly. 
# A better approach is usually to use a placeholder like "Unknown" or the mean/median for numbers.
```

```python
#Show the dataframe if we filled the null values for the fields cust_fname with "Dave" and account_type with "Not Available".
print("*** Filling Nulls with 'Dave' and 'Not Available' ***")
df.fillna({
    "cust_fname": "Dave", 
    "account_type": "Not Available"
}).show()
```

```python
#Show how the existing dataframe would look like with rows including null values being dropped.
print("*** Dropping Any Row with Null Values ***")
df.na.drop().show()
```

```python
#Show how the existing dataframe would look like with rows including null values for account_type' being dropped.
print("*** Dropping Rows where 'account_type' is Null ***")
df.na.drop(subset=["account_type"]).show()
```

```python
#Show how the existing dataframe would look like with rows including null values for 'cust_fname' being dropped.
print("*** Dropping Rows where 'cust_fname' is Null ***")
df.na.drop(subset=["cust_fname"]).show()
```

```python
#Show how the existing dataframe would look like if we replaced "Gabriella" with Gaby
print("*** Replacing 'Gabriella' with 'Gaby' ***")
df.replace("Gabriella", "Gaby").show()
```

```python
#Show how the existing dataframe would look like if we dropped the account_type column.
print("*** Dropping the 'account_type' Column ***")
df.drop("account_type").show()
```
